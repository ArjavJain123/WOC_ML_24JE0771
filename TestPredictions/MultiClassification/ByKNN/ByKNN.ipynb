2. Using Pandas to read the csv file

```python
data = pd.read_csv(csvOutputFile)
data.head()
```

```python
feature_col = [f"Feature_{i}" for i in range(1,21)]
X = data[feature_col].to_numpy()
Y = data['Class'].to_numpy()
IDs = data['ID'].to_numpy()
```

3. Standardizing X using z-score

```python
X_mean = np.mean(X, axis=0)
X_std = np.std(X, axis=0)
X = (X - X_mean) / X_std
```

4. Importing the test dataset 'multi_classification_test.csv'

```python
DataFileID = '15OBsD57pRHqrOPkWXbaeUs1kP6pZK9CN'
url = f'https://drive.google.com/uc?id={DataFileID}'
DataOutputFile = 'multi_classification_test.csv'
gdown.download(url, DataOutputFile, quiet=False)
df = pd.read_csv(DataOutputFile)
```

```python
print(df.head())
```

5. Standardizing X_data

```python
X_data = df[feature_col].to_numpy()
X_data = (X_data- X_mean) / X_std
print(X_data.shape)
```

6. Initializing the model and then generating predictions array using the X_data array. Using k=9, which I found to be the optimal k value, using elbow method (in experimentation files)

```python
model = KNN()
predictions = model.predict(X, Y, X_data, knn_k=9)
```

7. Finally, storing the target class labels vs IDs into pandas DataFrame and then exporting it to csv file format

```python
mappings = []
for i in range(X_data.shape[0]):
  mappings.append((i+1, predictions[i]))
OutDF = pd.DataFrame(mappings, columns=['ID', 'Class'])
OutDF.to_csv('output.csv', index=False)
```
