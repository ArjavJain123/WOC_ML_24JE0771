## MultiClassification Task Using NN algortish:

Importing data files and required modules:

```python
# from google.colab import drive
# drive.mount('/content/drive')
!pip install gdown
import gdown
#this way, anyone can access the files without mounting google driving. Accessing the csv files:
csvFileID = '1f9F5zZLY1nawh3E1Ay8LPlhLcFgKuvDt'
url = f'https://drive.google.com/uc?id={csvFileID}'
csvOutputFile = 'multi_classificatoin_train.csv'
gdown.download(url, csvOutputFile, quiet=False)

#accessing the module files:
moduleFileID = '1DemTzwacvp2wiKVlkuU9T5Vdw_d48Afh'
url = f'https://drive.google.com/uc?id={moduleFileID}'
moduleOutputFile = 'NeuralNetwork.py'
gdown.download(url, moduleOutputFile, quiet=False)
from NeuralNetwork import NeuralNetwork
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(csvOutputFile)
data.head()
```

```python
feature_col = [f"Feature_{i}" for i in range(1, 21)]
X = data[feature_col].to_numpy()
Y = data['Class'].to_numpy()
print(X.shape)
print(Y.shape)
```

Performing K fold CV, using the hyperparamaters and model architecture from experimentation files:

```python
l1 = [20, 16,16,5]
act_funcs1 = ['relu','relu','softmax']
model1 = NeuralNetwork(l1, act_funcs1, alpha=1e-3, epochs=500)
model1.k_fold_cv(X.T, Y, 5,cost_func='soft',details=False)
```

Now training on entire X.
X must be standardized before that. Using z score:

```python
X_mean = np.mean(X, axis=0)
X_std = np.std(X, axis=0)
X = (X - X_mean)/X_std
model1.train(X.T, Y, cost_func='soft', details=False, plot_costs=True)
```

```python
YFileID = '15OBsD57pRHqrOPkWXbaeUs1kP6pZK9CN'
url = f'https://drive.google.com/uc?id={YFileID}'
HiddenData = 'multi_classification_test.csv'
gdown.download(url, HiddenData, quiet=False)
```

```python
df = pd.read_csv(HiddenData)
```

```python
X_data = df[feature_col].to_numpy()
X_data = (X_data - X_mean)/X_std
```

Mapping and then exporting the target vs ID file

```python
mapping = []
predictions = model1.predict(X_data.T)
for i in range(X_data.shape[0]):
  mapping.append([i+1,predictions[i]])

new_df = pd.DataFrame(mapping, columns=['ID', 'Class'])
new_df.sort_values(by='ID', inplace=True)
print(new_df.head())
new_df.to_csv('multiclassMapping.csv', index=False)
```
