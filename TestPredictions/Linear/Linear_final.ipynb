# Dataset: linear_regression_train.csv

1. Importing csv files and modules:

```python
# from google.colab import drive
# drive.mount('/content/drive')
!pip install gdown
import gdown
#this way, anyone can access the files without mounting google driving. Accessing the csv files:
csvFileID = '1xg3qw9Pkh3flUAhN4fXgcv9YbmodFJZ-'
url = f'https://drive.google.com/uc?id={csvFileID}'
csvOutputFile = 'linear_regression_train.csv'
gdown.download(url, csvOutputFile, quiet=False)

#accessing the module files:
moduleFileID = '194P0NEQxWQOpYLYSknx3I2iudtESGPrQ'
url = f'https://drive.google.com/uc?id={moduleFileID}'
moduleOutputFile = 'LinearRegression.py'
gdown.download(url, moduleOutputFile, quiet=False)
from LinearRegression import LinearRegression
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

2. Reading csv files using pandas

```python
data = pd.read_csv(csvOutputFile)
data.head()
```

3. Extracting X and Y from the dataframe, and then z score standardizing X:

```python
feature_col = [f"Feature_{i}" for i in range(1,26)]
X = data[feature_col].to_numpy()
Y = data["Target"].to_numpy()
X_mean = np.mean(X, axis=0)
X_std = np.std(X, axis=0)
X = (X - X_mean) / X_std
```

4. Training the X on Linear Regression model with alpha = 3e-3, L2 regularization parameter (lambda) = 3e-3, number of epochs=10000.
The value of the hyperparameters are obtained by hyperparameter tuning, done in the experimentation files.

```python
model = LinearRegression(inp_features=25,epochs=10000,batch_size=X.shape[0],alpha=3e-3,reg_param=3e-3)
model.train(X, Y, details=False, plot_costs=True)
```

5. Importing the DataSet *'linear_regression_test.csv'* and then using Pandas to Read it

```python
dataFileID = '1_o58wjcBNVoYry5fsFZ59x3xgkm5m0GN'
url = f'https://drive.google.com/uc?id={dataFileID}'
DataOutFile = 'linear_regression_test.csv'
gdown.download(url, DataOutFile, quiet=False)
df = pd.read_csv(DataOutFile)
df.head()
```

6. Extracting the X_data from the dataframe and then standardizing X_data (z-score)

```python
X_data = df[feature_col].to_numpy()
X_data = (X_data - X_mean) / X_std
```

7. Using the Tranined Model to make predictions,and then exporting predictions

```python
predictions = model.predict(X_data)
mappings = []
for i in range(predictions.shape[0]):
  mappings.append((i+1, predictions[i]))

outDataFrame = pd.DataFrame(mappings, columns=["Id", "Target"])
outDataFrame.to_csv("predictions.csv", index=False)
```
